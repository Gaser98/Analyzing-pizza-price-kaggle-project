```python
#Q4: Suppose the Pizza Hut company owns P={X|X unique price value} and Domnino's Pizza owns D={X|X unique price value} what is P∩D,P∪D?
price_pizzahut=set(df.loc[df['Company']=='Pizza Hut']['Price'].to_list())
price_dominos=set(df.loc[df['company']=='Domino s Pizza']['Price'].to_list())
```

```python
#Q6: In a city, the city has 4 pizza selling companies, namely Domino's Pizza, Pizza Hut, Godfather's Pizza, and IMO's Pizza, Employees want to buy pizza to accompany their lunch break, unfortunately these employees can only choose to buy from 2 companies only, How many choices can the five employees choose?
#In other terms,How many cominations of 2 companies can be found?
import itertools
companies = ["Domino's Pizza", "Pizza Hut", "Godfather's Pizza", "IMO's Pizza"]
combination, i = itertools.combinations(companies, 2), 1
for comb in combination:
    print(f"combination {i}: {comb}")
    i += 1
```

```python
#6 combinations
```
